```python
html_content = "jumia-flash-sales.html"
with open(html_content, 'r', encoding='utf-8') as file:
    html_content = file.read()
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
print(soup)
```

```python
containers = soup.find_all('div', class_='prd card _full')
script_tag = soup.find('script', string=re.compile(r'dataLayer'))

sku_match = re.search(r'"skus": \[(.*?)\]', script_tag.string)
```

```python
for container in containers:
    
    skus = sku_match.group(1).replace('"', '').split(', ') if sku_match else []

    item_name_element = container.find('h3', class_='name')

    item_name = item_name_element.text.strip() if item_name_element else "N/A"

    current_price_element = container.find('span', class_='curr')

    current_price = current_price_element.text.strip() if current_price_element else "N/A"

    old_price_element = container.find('span', class_='old')

    old_price = old_price_element.text.strip() if old_price_element else "N/A"
```
